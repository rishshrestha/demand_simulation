```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
pop = 100
annual_growth_rate = .10
p_adopt_connection = .7
p_disconnect_connection = .01
pop_testing = []

def annual_no_of_load_growth(pop):
    """Run the simulation."""
    for _ in range(0,365):
        global annual_growth_rate,p_taking_connection,p_disconnect_connection,pop_testing
        enter_village = 1*(np.random.rand() <= (annual_growth_rate * pop/365))
        if enter_village ==1:
            adopt_connection = 1*(np.random.rand() <= (p_adopt_connection*pop/365))
        else: adopt_connection = 0
        disconnect_connection = 1*(np.random.rand() <= p_disconnect_connection*pop/365)
        pop += adopt_connection - disconnect_connection
    
    pop_testing.append(pop)
    return pop

bins = np.linspace(0, pop+50, 25);
for i, nsteps in enumerate(range(0,10)):
    annual_no_of_load_growth(pop)
pd.DataFrame(pop_testing).hist(edgecolor='k')
plt.xlabel("Population size");
plt.ylabel("Histogram");
plt.title("{0:d} simulation".format(10));
pop=100
for i, nsteps in enumerate(range(0,100)):
    annual_no_of_load_growth(pop)
pd.DataFrame(pop_testing).hist(edgecolor='k')
plt.xlabel("Population size");
plt.title("{0:d} simulation".format(100));
pop=100
for i, nsteps in enumerate(range(0,1000)):
    annual_no_of_load_growth(pop)
pd.DataFrame(pop_testing).hist(edgecolor='k')
plt.xlabel("Population size");
plt.title("{0:d} simulation".format(1000));
```
